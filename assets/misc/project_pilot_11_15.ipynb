```python
import pandas as pd
#pd.set_option('display.max_columns', None)
pd.reset_option(pat="all") 
import matplotlib.pyplot as plt
df=pd.read_csv("SCDB_2019_01_justiceCentered_Vote.csv",encoding = 'unicode_escape')
print(df.shape)
print(df.head)
print(df.dtypes)
```

```python
RBG_df=df[df['justiceName']=="RBGinsburg"]
print(RBG_df)
```

```python
#issueArea
#issue
issue_grouped_df=df.groupby("issue")
print(issue_grouped_df.size().sort_values(ascending=False))
```

```python
#count only one for the same cases together and count each case only one time when analyzing the issue that occurs the most
pd.reset_option(pat="all") 
unique_df = df.drop_duplicates('caseId')
print(unique_df)
print(unique_df.groupby('term').size())
plt.plot(unique_df.groupby('term').size())
plt.title("Supreme Court Cases by Year")
plt.xlabel('Year')
plt.ylabel('Number of Cases taken')
plt.show()
print("aggregated by term and issue")
pd.set_option('display.max_rows', None)
issue_grouped_df=unique_df.groupby(["term","issue"]).agg({'caseId': 'count'})
issue_grouped_df=issue_grouped_df.sort_values(by=["term","caseId"],ascending=False)

print(issue_grouped_df.head)
print()
print("aggregated by term, issue and winning party")
issue_grouped_df=unique_df.groupby(["term","issue","partyWinning"]).agg({'caseId': 'count'})
print(issue_grouped_df.sort_values(by=["term","caseId"],ascending=False)) #some apparent winning patterns shown in comparison 


```

```python
print()
print("aggregated by issue, term, and winning party")
issue_grouped_df=unique_df.groupby(["issue","partyWinning"]).agg({'caseId': 'count'})
print(issue_grouped_df.sort_values(by=["issue"],ascending=False)) #some apparent winning patterns shown in comparison 

```

```python
#Let's have a look at what type/issue of cases RBG has the strongest opinions about
RBG_df=df[(df['justiceName']=="RBGinsburg")].drop_duplicates('caseId')
RBG_df=RBG_df[(RBG_df['opinion']==2) | (RBG_df['opinion']==3)]
issue_grouped_df=RBG_df.groupby(["term","issue"]).agg({'caseId': 'count'})
issue_grouped_df=issue_grouped_df.sort_values(by=["term","caseId"],ascending=False)

print(issue_grouped_df)
print()
print(issue_grouped_df.shape)
```

```python
RBG_strong_opinion=RBG_df[(RBG_df['term']==2016) & ((RBG_df['opinion']==2) | (RBG_df['opinion']==3))]

print(RBG_strong_opinion['caseName'])
print("decision date")
print(RBG_strong_opinion['dateDecision'])
print("print usCite which is the citation to each case from the official United States Reports")
print(RBG_strong_opinion['usCite'])
print("print usCite which is the citation to each case from the Supreme Court Reporter (S.CT)")
print(RBG_strong_opinion['sctCite'])
print("print usCite which is the citation to each case from the Lawyers' Edition of the United States Reports(LEd)")
print(RBG_strong_opinion['ledCite'])
print("print usCite which is the citation to each case from the LEXIS cite. ")
print(RBG_strong_opinion['lexisCite'])
```
