```python
"""
    Leetcode 1038. Binary Search Tree to Greater Sum Tree
"""
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, x):
#         self.val = x
#         self.left = None
#         self.right = None

class Solution:
    sum = 0
    def bstToGst(self, root: TreeNode) -> TreeNode:
        def reverseOrderTraversal(node: TreeNode):
            if node.right:
                reverseOrderTraversal(node.right)
            self.sum += node.val
            node.val = self.sum
            if node.left:
                reverseOrderTraversal(node.left)
        reverseOrderTraversal(root)
        return root
        
```

```python
"""
Leetcode 1214 Two Sum BSTs

Given two binary search trees, return True if and only if there is a node in the first tree and a node in the second 
tree whose values sum up to a given integer target.
"""
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, x):
#         self.val = x
#         self.left = None
#         self.right = None


#solution1-recursion
class Solution:
    def twoSumBSTs(self, root1: TreeNode, root2: TreeNode, target: int) -> bool:
        def helper(root1,root2,target):
            if not root1 or not root2:
                return False
            if root1.val+root2.val==target:
                return True
            if root1.val+root2.val<target:
                return helper(root1,root2.right,target) or helper(root1.right,root2,target)
            if root1.val+root2.val>target:
                return helper(root1,root2.left,target) or helper(root1.left,root2,target)
        return helper(root1,root2,target)

#solution 2 - tree traversal and hashmap
class Solution:
    def twoSumBSTs(self, root1: TreeNode, root2: TreeNode, target: int) -> bool:
        def inOrder(root):
            if not root:
                return []
            lst=[]
            return inOrder(root.left)+[target-root.val]+inOrder(root.right)
        lst=inOrder(root1) #O(n)
        def traverse(root,target):
            if not root:
                return False
            if root.val in target:
                return True
            return traverse(root.left,lst) or traverse(root.right,lst)
        return traverse(root2,lst)
```

```python
"""
#9 HARD
This problem was asked by Airbnb.

Given a list of integers, write a function that returns the largest sum of non-adjacent numbers. Numbers can be 0 or 
negative.

For example, [2, 4, 6, 2, 5] should return 13, since we pick 2, 6, and 5. [5, 1, 1, 5] should return 10, since we pick
5 and 5.

Follow-up: Can you do this in O(N) time and constant space?
"""
def largest_sum(lst):
    if not lst:
        return None
    if len(lst)==0:
        return 0
    if len(lst)<2:
        return lst[0]
    if len(lst)==2:
        return max(lst[0],lst[1])
    #+2 or +3 each time
    p1=0
    num_sum1=lst[0]
    while p1<len(lst):
        if p1+3<len(lst):
            if num_sum1+lst[p1+2]>num_sum1+lst[p1+3]:
                num_sum1+=lst[p1+2]
                p1+=2
            else:
                num_sum1+=lst[p1+3]
                p1+=3
        elif p1+2<len(lst):
            num_sum1+=lst[p1+2]
            p1+=2
    p2=1
    num_sum2=lst[1]
    while p2<len(lst):
        if p2+3<len(lst):
            if num_sum2+lst[p2+2]>num_sum2+lst[p2+3]:
                num_sum2+=lst[p2+2]
                p2+=2
            else:
                num_sum2+=lst[p2+3]
                p2+=3
        elif p2+2<len(lst):
            num_sum2+=lst[p2+2]
            p1+=2
    return max(num_sum1,num_sum2)
print(largest_sum([2, 4, 6, 2, 5]))
print(largest_sum([5, 1, 1, 5]))
```

```python
"""
#12
This problem was asked by Amazon.

There exists a staircase with N steps, and you can climb up either 1 or 2 steps at a time. Given N, write a function 
that returns the number of unique ways you can climb the staircase. The order of the steps matters.

For example, if N is 4, then there are 5 unique ways:

    1, 1, 1, 1
    2, 1, 1
    1, 2, 1
    1, 1, 2
    2, 2

What if, instead of being able to climb 1 or 2 steps at a time, you could climb any number from a set of positive 
integers X? For example, if X = {1, 3, 5}, you could climb 1, 3, or 5 steps at a time.

"""

def ways_of_stair(N):
    if N<=0:
        return 1
    if N==1:
        return 1
    if N==2:
        return 2
    return ways_of_stair(N-1)+ways_of_stair(N-2)


print(ways_of_stair(4))


def three_way_of_stair(N):
    if N<=0:
        return 1
    if N==1:
        return 1
    if N==3:
        return 2
    return three_way_of_stair(N-1)+three_way_of_stair(N-3)+three_way_of_stair(N-5)


def staircase(n, X):
    cache = [0 for _ in range(n + 1)]
    cache[0] = 1
    for i in range(n + 1):
        cache[i] += sum(cache[i - x] for x in X if i - x > 0)
        cache[i] += 1 if i in X else 0
    return cache[-1]

print(three_way_of_stair(5))
```
