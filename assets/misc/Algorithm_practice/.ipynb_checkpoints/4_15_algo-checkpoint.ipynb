```python
"""
593. Valid Square
Given the coordinates of four points in 2D space, return whether the four points could construct a square.

The coordinate (x,y) of a point is represented by an integer array with two integers.

Example:

Input: p1 = [0,0], p2 = [1,1], p3 = [1,0], p4 = [0,1]
Output: True

 

Note:

    All the input integers are in the range [-10000, 10000].
    A valid square has four equal sides with positive length and four equal angles (90-degree angles).
    Input points have no order.

"""

def validSquare(p1,p2,p3,p4) -> bool:
    def distance(p1,p2):
        return (p1[0]-p2[0])**2+(p1[1]-p2[1])**2
    lst=[]
    lst.append(distance(p1,p2))
    lst.append(distance(p1,p3))
    lst.append(distance(p1,p4))
    lst.append(distance(p2,p3))
    lst.append(distance(p2,p4))
    lst.append(distance(p3,p4))
        #print(lst)
    max_dis=max(lst)
        #print(max_dis)
    min_dis=min(lst)
        #print(min_dis)
    count_max=0
    count_min=0
    for i in lst:
        if i!=max_dis and i!=min_dis:
            return False
        if i==max_dis:
            count_max+=1
        if i==min_dis:
            count_min+=1
    return max_dis==2*min_dis and count_max==2 and count_min==4
        
        
p1 = [0,0]
p2 = [1,1]
p3 = [1,0]
p4 = [0,1]

print(validSquare(p1,p2,p3,p4)==True)
```

```python
"""
661. Image Smoother
Given a 2D integer matrix M representing the gray scale of an image, you need to design a smoother to make the gray scale of each cell becomes the average gray scale (rounding down) of all the 8 surrounding cells and itself. If a cell has less than 8 surrounding cells, then use as many as you can.

Example 1:

Input:
[[1,1,1],
 [1,0,1],
 [1,1,1]]
Output:
[[0, 0, 0],
 [0, 0, 0],
 [0, 0, 0]]
Explanation:
For the point (0,0), (0,2), (2,0), (2,2): floor(3/4) = floor(0.75) = 0
For the point (0,1), (1,0), (1,2), (2,1): floor(5/6) = floor(0.83333333) = 0
For the point (1,1): floor(8/9) = floor(0.88888889) = 0

Note:

    The value in the given matrix is in the range of [0, 255].
    The length and width of the given matrix are in the range of [1, 150].

"""

def imageSmoother(M):
    R, C = len(M), len(M[0])
    ans = [[0] * C for _ in M]
    for r in range(R):
        for c in range(C):
            count=0
            for nr in (r-1,r,r+1):
                for nc in (c-1,c,c+1):
                    if 0<=nr<R and 0<=nc<C:
                        ans[r][c]+=M[nr][nc]
                        count+=1
            ans[r][c]=ans[r][c]//count
                
    return ans

inp=[[1,1,1],[1,0,1],[1,1,1]]
print(imageSmoother(inp))
```

```python
"""
75. Sort Colors
Given an array with n objects colored red, white or blue, sort them in-place so that objects of the same color are adjacent, with the colors in the order red, white and blue.

Here, we will use the integers 0, 1, and 2 to represent the color red, white, and blue respectively.

Note: You are not suppose to use the library's sort function for this problem.

Example:

Input: [2,0,2,1,1,0]
Output: [0,0,1,1,2,2]

Follow up:

    A rather straight forward solution is a two-pass algorithm using counting sort.
    First, iterate the array counting number of 0's, 1's, and 2's, then overwrite array with total number of 0's, then 1's and followed by 2's.
    Could you come up with a one-pass algorithm using only constant space?


"""
#solution 1 - 3 pass, one for each color
def sortColor(nums):
    color_list=[0,1,2]
    def sort(nums,color):
        p1=0
        p2=len(nums)-1
        while p1<p2:
            if nums[p1]>color and nums[p2]==color:
                temp=nums[p1]
                nums[p1]=nums[p2]
                nums[p2]=temp
                p1+=1
                p2-=1
            if nums[p1]<=color:
                p1+=1
            if nums[p2]!=color:
                p2-=1 
        return nums
    for i in color_list:
        nums=sort(nums,i)
    return nums


#Solution 2- 1 pass
def sortColor(nums):
    color_list=[0,1,2]
    def sort(index,nums,color):
        p1=index
        p2=len(nums)-1
        while p1<p2:
            if nums[p1]>color and nums[p2]==color:
                temp=nums[p1]
                nums[p1]=nums[p2]
                nums[p2]=temp
                p1+=1
                p2-=1
            if nums[p1]<=color:
                p1+=1
            if nums[p2]!=color:
                p2-=1 
        return nums,index
        #index=0
    for i in color_list:
        nums,index=sort(0,nums,i)
        """
        OR
        nums,index=sort(index,nums,0)
        nums,index=sort(index,nums,1)
        nums,index=sort(index,nums,2)
        """
    return nums
lst=[2,0,2,1,1,0]
print(sortColor(lst)==[0,0,1,1,2,2])
```
