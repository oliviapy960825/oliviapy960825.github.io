# Task 3
Time the difference between processing a ufunc/broadcasting compared to writing a for loop to iterate over a numpy array

**HINT: You'll want to use `for x in np.nditer(arr1, op_flags=['readwrite']):` to be able to iterate and update the elements in our numpy array**

```python
import time

class TimeTracker():
    def __init__(self):
        pass
    
    def start(self):
        self.start_time = time.time()
        
    def stop(self):
        self.stop_time = time.time()
        
    def run(self, func, params):
        self.start()
        func(*params)
        self.stop()
        return self.runtime()
        
    def runtime(self):
        return f"Runtime: {self.stop_time - self.start_time} seconds"
```

```python
tracker = TimeTracker()

tracker.start()
for ix in range(100000000):
    pass
tracker.stop()
print(tracker.runtime())
```

```python
import numpy as np
"""
will return nothing but print out the runtime of a for loop and a ufunc
"""
arr = np.arange(1000000)
print(arr)
tracker = TimeTracker()

tracker.start()
# Place to put code for iterative loop
#with np.nditer(arr, op_flags=['readwrite']) as it:
#    for x in it:
for x in np.nditer(arr, op_flags=['readwrite']):
    x=x**2
print(arr)
tracker.stop()
print(tracker.runtime())

tracker.start()
# Place for ufunc or vectorized operation
np.square(arr)
tracker.stop()
print(tracker.runtime())
```

## Task 4
Create a random array of integers, from -20 to 20, and determine how many of the elements are positive using **np.where()**.

```python
# Space for Task 4
import numpy as np
import numpy.random as rand
"""
will return nothing but print out the number of elements that are positive in a random array
of integers from -20 to 20
First of all it will generate a random arrat of integers from -20 to 20, and then it will determine
how many positive elements there are in the array and then count the number of the positive elements
in the array
"""
np.random.seed(1)
arr=rand.randint(-20,20,size=(4,5))
print(arr)
print(np.where(arr>0,1,0))
sum=0
for i in np.where(arr>0,1,0):
    for j in i:
        sum+=j
print(sum)
```
