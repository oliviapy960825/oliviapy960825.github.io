```python
class TwitterStatus():
    """
    TwitterStatus class has attributes: created_date, description, favorites_count, friends_count,
    id_str, listed_count, location, name, screen_name, statuses_count, url, verified, hashtags, user_mentions,
    retweet_count and student_name
    TwitterStatus class has a method: to_json, which includes every attribute of TwitterStatus class and make
    a json format string and return the json string
    """
    def __init__(self, item):
        """
        This is the initialization method of the TwitterStatus class
        Args:
        self, item
        Returns:
        none
        """
        self.created_date=item['created_at']
        self.description=item['user'].get('description')
        self.favorites_count=str(item['user'].get('favorites_count'))
        self.friends_count=str(item['user'].get('friends_count'))
        self.id_str=str(item['id_str'])
        self.listed_count=str(item['user'].get('listed_count'))
        self.location=item['user'].get('location')
        self.name=item['user'].get('name')
        self.screen_name=item['user'].get('screen_name')
        self.statuses_count=str(item['user'].get('statuses_count'))
        self.url=item['user'].get('url')
        self.verified=item['user'].get('verified')
        self.hashtags=item['hashtags']
        self.user_mentions=str(item['user_mentions'])
        self.retweet_count=str(item['retweet_count'])
        self.student_name="Wang Peiyu"
        
        
    def to_json(self):
        """
        This method takes the attrbiutes of TwitterStatus class and return a json format string
        Args:
        self
        return:
        json_string(string)-json format string
        """
        json_string={
                'created_date':self.created_date,
                'description':self.description,
                'favorites_count':self.favorites_count,
                'friends_count':self.friends_count,
                'id_str':self.id_str,
                'listed_count':self.listed_count,
                'location':self.location,
                'name':self.name,
                'screen_name':self.screen_name,
                'statuses_count':self.statuses_count,
                'url':self.url,
                'verified':self.verified,
                'hashtags':self.hashtags,
                'user_mention':self.user_mentions,
                'student_name':self.student_name
            }
        return (json_string)
```
