### 2.(b). Implement the Recognizing Hand-written digits tutorial

```python
print(__doc__)

# Author: Gael Varoquaux <gael dot varoquaux at normalesup dot org>
# License: BSD 3 clause

# Standard scientific Python imports
import matplotlib.pyplot as plt

# Import datasets, classifiers and performance metrics
from sklearn import datasets, svm, metrics

# The digits dataset
digits = datasets.load_digits() #load digits dataset from datasets into digits

# The data that we are interested in is made of 8x8 images of digits, let's
# have a look at the first 4 images, stored in the `images` attribute of the
# dataset.  If we were working from image files, we could load them using
# matplotlib.pyplot.imread.  Note that each image must have the same size. For these
# images, we know which digit they represent: it is given in the 'target' of
# the dataset.
images_and_labels = list(zip(digits.images, digits.target)) #put the first 4 images of the digits data and the real number of the digits into a container and convert it into a list called images_and_labels
for index, (image, label) in enumerate(images_and_labels[:4]): #for loop till the 5th element of images_and_labels list
    plt.subplot(2, 4, index + 1)#create a two rows 4 columns plot with indexes for the first row
    plt.axis('off')#set the axis of the plot to off
    plt.imshow(image, cmap=plt.cm.gray_r, interpolation='nearest') #display the image represented by the "image" variable
    plt.title('Training: %i' % label) #add a title to the plot

# To apply a classifier on this data, we need to flatten the image, to
# turn the data in a (samples, feature) matrix:
n_samples = len(digits.images) #extract the size/length of digits.images and save it to n_samples
data = digits.images.reshape((n_samples, -1)) #reshape the data stored in data.images into a (samples, feature) matrix and save the results to variable data

# Create a classifier: a support vector classifier
classifier = svm.SVC(gamma=0.001) #black box still

# We learn the digits on the first half of the digits
classifier.fit(data[:n_samples // 2], digits.target[:n_samples // 2]) # fit/train the classifier according to the first half of the digits data and target (the real number the data represents)

# Now predict the value of the digit on the second half:
expected = digits.target[n_samples // 2:] #extract the second half of the numbers the data represents from digits.target
predicted = classifier.predict(data[n_samples // 2:]) # use the trained classifier and the second half set of the digits data to predict the second half of numbers

print("Classification report for classifier %s:\n%s\n"
      % (classifier, metrics.classification_report(expected, predicted))) #print the classifier and the metrix of the real numbers in the data set in contrast of the numbers predicted by the trained classifier
print("Confusion matrix:\n%s" % metrics.confusion_matrix(expected, predicted)) # print the metrix of the real numbers in the data set in contrast of the numbers predicted by the trained classifier

images_and_predictions = list(zip(digits.images[n_samples // 2:], predicted)) # put the second half of the digits images and the predicted numbers by the classifier into a container and convert it into a list called images_and_predictions
for index, (image, prediction) in enumerate(images_and_predictions[:4]):#for loop till the 5th element of images_and_predictions list
    plt.subplot(2, 4, index + 5)#create a two rows 4 columns plot with indexes for the second row
    plt.axis('off')#set the axis of the plot to off
    plt.imshow(image, cmap=plt.cm.gray_r, interpolation='nearest')#display the image represented by the "image" variable
    plt.title('Prediction: %i' % prediction)#add a title to the plot

plt.show()
```
