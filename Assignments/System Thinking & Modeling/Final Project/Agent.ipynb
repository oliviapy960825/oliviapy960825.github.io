```python
#packages and global variables
import math
import numpy
from mesa import Agent, Model
from pprint import pprint
healthy_to_depressed_prob=
healthy_to_obese_prob=
depressed_to_obese_prob=
obese_to_depressed_prob=
depressed_seeking_help_prob=0.33
depressed_sticking_to_treatment_prob=0.21
obese_seeking_help_prob=0.667
obese_sticking_to_treatment_prob=0.483
obese_healthy_prob=0.09
```

```python

class Healthy(Agent):
    """
    This is the class of healthy population who are neither depressed nor obesed and the
    starting point and base of our model
    
    attribute:
    unique_id: unique int
    depress_flag: int, value of 1 or 0, indicating if a person is depressed or not
    obese_flag: int, value of 1 or 0, indicating if a person is obese or not
    healthy_to_depressed_prob: float, value of the probability of healthy person becoming depressed
    healthy_to_obese_prob: float, value of the probability of healthy person becoming obese
    """
    def __init__(self, unique_id, model, depressed_flag, obese_flag):
        """
        There are six attributes for a healthy agent:
        unique_id: the unique int type id of a healthy agent
        model: the model the agent is associated to
        depressed_flag: the int value(0 or 1) to indicate that if this agent is depressed
        obese_flag: the int value(0 or 1) to indicate that if this agent is obese
        healthy_to_depressed_prob: the float type value of the probability that a healthy agent falling depressed
        healthy_to_obese_prob: the float type value of the probability that a healthy agent falling obese
        """
        super().__init__(unique_id, model)
        #self.breed = 'healthy_people'
        self.depressed_flag=depressed_flag
        self.obese_flag=obese_flag
    def step(self):
        """
        This is the function which would randomly decide if a person is going to be depressed or obese or both based
        on the probabilities. If so then it will return the corresponding new type of this person, if not then it will
        return its old self
        """
        self.depressed_flag=numpy.random.choice(numpy.arange(0,2), p=[(1-healthy_to_depressed_prob), healthy_to_depressed_prob])
        self.obese_flag=numpy.random.choice(numpy.arange(0,2), p=[(1-healthy_to_obese_prob),healthy_to_obese_prob])
        if self.depressed_flag and self.obese_flag:#depressed and obese
            self.depressed_seeking_help_flag=0
            self.obese_seeking_help_flag=0
            return Depressed_Obese(self.unique_id,self.model,self.depressed_flag,self.obese_flag,self.depressed_seeking_help_flag, self.obese_seeking_help_flag)
        if self.depressed_flag and (not self.obese_flag):#depressed but not obese
            self.depressed_seeking_help_flag=0
            return Depressed(self.unique_id,self.model,self.depressed_flag,self.obese_flag,self.depressed_seeking_help_flag)
        if self.obese_flag and (not self.depressed_flag):#obese but not depressed
            self.obese_seeking_help_flag=0
            return Obese(self.unique_id,self.model,self.depressed_flag,self.obese_flag,self.obese_seeking_help_flag)
        else:#not obese and not depressed--old self
            return self
```

```python
class Depressed(Agent):
    """
    This is the class of depressed population, a depressed person can seek help (or not) and/or becoming obese 
    
    attribute:
    unique_id: unique int
    depress_flag: int, value of 1 or 0, indicating if a person is depressed or not
    obese_flag: int, value of 1 or 0, indicating if a person is obese or not
    depressed_seeking_help_flag: int, value of 1 or 0, indicating if this depressed person is seeking help
    """
    def __init__(self, unique_id, model, depressed_flag,obese_flag,depressed_seeking_help_flag):
        """
        This is the constructor of depressed agent class
        """
        super().__init__(unique_id, model)
        #self.breed = 'depressed_people'
        self.depressed_flag=depressed_flag
        self.obese_flag=obese_flag
        self.depressed_seeking_help_flag=depressed_seeking_help_flag
        
    def step(self):
        """
        This is the function which would randomly decide if a person who is depressed is going to seeking help (or not)
        and/or become obese based on the probabilities. If so then it will return the corresponding new type of this 
        person, if not then it will return its old self
        """
        self.depressed_seeking_help_flag=numpy.random.choice(numpy.arange(0,2), p=[(1-depressed_seeking_help_prob), depressed_seeking_help_prob])
        self.obese_flag=numpy.random.choice(numpy.arange(0,2), p=[(1-depressed_to_obese_prob), depressed_to_obese_prob])
        if self.depressed_seeking_help_flag and self.obese_flag: #depressed and seeking help, and obese (and not seeking help, yet)
            depressed_sticking_to_treatment_flag=0
            obese_sticking_to_treatment_flag=0
            obese_seeking_help_flag=0
            return Depressed_Obese_Seeking_Help(self.unique_id, self.model, self.depressed_flag, self.obese_flag, self.depressed_seeking_help_flag, obese_seeking_help_flag, depressed_sticking_to_treatment_flag, obese_sticking_to_treatment_flag)
        if self.depressed_seeking_help_flag and (not self.obese_flag):#still depressed and not obese but started seeking help
            depressed_sticking_with_treatment_flag=0
            return Depressed_Seeking_Help(self.unique_id, self.model, self.depressed_flag, self.obese_flag,depressed_sticking_with_treatment_flag)
        if (not self.depressed_seeking_help_flag) and self.obese_flag:#obese and depressed but not seeking help yet
            obese_seeking_help_flag=0
            return Depressed_Obese(self.unique_id, self.model, self.depressed_flag,self.obese_flag,self.depressed_seeking_help_flag, obese_seeking_help_flag)
        if (not self.depressed_seeking_help_flag) and (not self.obese_flag):#still depressed and not seeking help and not obese
            return self
```

```python
class Depressed_Seeking_Help(Agent):
    """
    This is the class of depressed and seeking help population, a depressed and seeking help person can stick to
    treatment (or not) and/or becoming obese 
    
    attribute:
    unique_id: unique int
    depress_flag: int, value of 1 or 0, indicating if a person is depressed or not
    obese_flag: int, value of 1 or 0, indicating if a person is obese or not
    depressed_sticking_with_treatment_flag: int, value of 1 or 0, indicating if this depressed person sticks with treatment or not
    """
    def __init__(self, unique_id, model, depressed_flag, obese_flag,depressed_sticking_to_treatment_flag):
        """
        
        """
        super().__init__(unique_id, model)
        #self.breed = 'depressed_people'
        self.depressed_flag=depressed_flag
        self.obese_flag=obese_flag
        self.depressed_sticking_to_treatment_flag=depressed_sticking_to_treatment_flag
    def step(self):
        """
        This is the function which would randomly decide if a person who is depressed and seeking help is going to 
        stick to treatment (or not) and/or become obese based on the probabilities. If so then it will return the 
        corresponding new type of this person, if not then it will return its old self
        """
        self.depressed_sticking_to_treatment_flag=numpy.random.choice(numpy.arange(0,2), p=[(1-depressed_sticking_to_treatment_prob), depressed_sticking_to_treatment_prob])
        self.obese_flag=numpy.random.choice(numpy.arange(0,2), p=[(1-depressed_to_obese_prob), depressed_to_obese_prob])
        if (not self.depressed_sticking_to_treatment_flag) and self.obese_flag:#obese and still depressed, just not seeking help anymore, and not seeking help with obesity
            depressed_seeking_help_flag=0
            obese_seeking_help_flag=0
            return Depressed_Obese(self.unique_id, self.model, self.depressed_flag,self.obese_flag,depressed_seeking_help_flag, obese_seeking_help_flag)
        if (not self.depressed_sticking_to_treatment_flag) and (not self.obese_flag): #still depressed but not obese, just not seeking treatment anymore
            depressed_seeking_help_flag=0
            return Depressed(self.unique_id, self.model, self.depressed_flag,self.obese_flag,depressed_seeking_help_flag)
        if self.depressed_sticking_to_treatment_flag and self.obese_flag:#depressed but seeking help, and obese now
            depressed_seeking_help_flag=1
            obese_seeking_help_flag=0
            obese_sticking_to_treatment_flag=0
            return Depressed_Obese_Seeking_Help(self.unique_id, self.model, self.depressed_flag, self.obese_flag, depressed_seeking_help_flag, obese_seeking_help_flag, self.depressed_sticking_to_treatment_flag, obese_sticking_to_treatment_flag)
        if self.depressed_sticking_to_treatment_flag and (not self.obese_flag):#still seeking treatment and not obese
            return self
```

```python
class Obese(Agent):
    """
     This is the class of obese population, an obese person can seek help (or not) and/or become depressed
    
    attribute:
    unique_id: unique int
    depress_flag: int, value of 1 or 0, indicating if a person is depressed or not
    obese_flag: int, value of 1 or 0, indicating if a person is obese or not
    obese_seeking_help_flag: int, value of 1 or 0, indicating if this obese person is seeking help
    """
    def __init__(self, unique_id, model, depressed_flag, obese_flag, obese_seeking_help_flag):
        """
        
        """
        super().__init__(unique_id, model)
        #self.breed = 'obese_people'
        self.depressed_flag=depressed_flag
        self.obese_flag=obese_flag
        self.obese_seeking_help_flag=obese_seeking_help_flag

    
    def step(self):
        """
        This is the function which would randomly decide if an obese person is going to seeking help (or not)
        and/or become depressed based on the probabilities. If so then it will return the corresponding new type of this 
        person, if not then it will return its old self
        """
        self.depressed_flag=numpy.random.choice(numpy.arange(0,2), p=[(1-obese_to_depressed_prob), obese_to_depressed_prob])
        self.obese_seeking_help_flag=numpy.random.choice(numpy.arange(0,2), p=[(1-obese_seeking_help_prob), obese_seeking_help_prob])
        if self.depressed_flag and self.obese_seeking_help_flag:#now depressed and obese (and seeking help)
            depressed_sticking_to_treatment_flag=0
            depressed_seeking_help_flag=0
            obese_sticking_to_treatment_flag=0
            return Depressed_Obese_Seeking_Help(self.unique_id, self.model, self.depressed_flag, self.obese_flag, depressed_seeking_help_flag, self.obese_seeking_help_flag, depressed_sticking_to_treatment_flag, obese_sticking_to_treatment_flag)
        if self.depressed_flag and (not self.obese_seeking_help_flag):#obese and depressed and not seeking help
            depressed_seeking_help_flag=0
            return Depressed_Obese(self.unique_id, self.model, self.depressed_flag,self.obese_flag,depressed_seeking_help_flag, self.obese_seeking_help_flag)
        if (not self.depressed_flag) and self.obese_seeking_help_flag:# not depressed but obese (but seeking help):
            obese_sticking_to_treatment_flag=0
            return Obese_Seeking_Help(self.unique_id, self.model, self.depressed_flag, self.obese_flag,obese_sticking_to_treatment_flag)
        if (not self.depressed_flag) and (not self.obese_seeking_help_flag):#not depressed but still obese and not seeking help
            return self
```

```python
class Obese_Seeking_Help(Agent):
    """
    This is the class of obese and seeking help population, an obese person and seeking help can stick to treatment
    (or not) and/or become depressed
    
    attribute:
    unique_id: unique int
    depress_flag: int, value of 1 or 0, indicating if a person is depressed or not
    obese_flag: int, value of 1 or 0, indicating if a person is obese or not
    obese_sticking_to_treatment_flag: int, value of 1 or 0, indicating if this obese person is sticking to treatment
    """
    def __init__(self, unique_id, model, depressed_flag, obese_flag,obese_sticking_to_treatment_flag):
        """
        
        """
        super().__init__(unique_id, model)
        #self.breed = 'obese_people'
        self.depressed_flag=depressed_flag
        self.obese_flag=obese_flag
        self.obese_sticking_to_treatment_flag=obese_sticking_to_treatment_flag
        
    def step(self):
        """
        This is the function which would randomly decide if an obese and seeking help person is going to stick to 
        treatment (or not) and/or become depressed based on the probabilities. If so then it will return the 
        corresponding new type of this person, if not then it will return its old self
        """
        self.obese_sticking_to_treatment_flag=numpy.random.choice(numpy.arange(0,2), p=[(1-obese_sticking_to_treatment_prob), obese_sticking_to_treatment_prob])
        if(self.obese_sticking_to_treatment_flag):
            self.obese_flag=numpy.random.choice(numpy.arange(0,2), p=[obese_healthy_prob, (1-obese_healthy_prob)])
        else:
            self.obese_flag=1
        self.depressed_flag=numpy.random.choice(numpy.arange(0,2), p=[(1-obese_to_depressed_prob), obese_to_depressed_prob])
        if self.obese_flag and self.obese_sticking_to_treatment_flag and (not self.depressed_flag):#this person is obese (but sticking to treatment) and not depressed
            return self
        if self.obese_flag and self.obese_sticking_to_treatment_flag and self.depressed_flag:#obese(but sticking to treatment) and depressed
            depressed_sticking_to_treatment_flag=0
            depressed_seeking_help_flag=0
            obese_seeking_help_flag=1
            return Depressed_Obese_Seeking_Help(self.unique_id, self.model, self.depressed_flag, self.obese_flag, depressed_seeking_help_flag, obese_seeking_help_flag, depressed_sticking_to_treatment_flag, self.obese_sticking_to_treatment_flag)
        if self.obese_flag and (not self.obese_sticking_to_treatment_flag) and self.depressed_flag:#obese and depressed and not seeking help
            depressed_seeking_help_flag=0
            obese_seeking_help_flag=0
            return Depressed_Obese(self.unique_id, self.model, self.depressed_flag,self.obese_flag,depressed_seeking_help_flag, obese_seeking_help_flag)
        if self.obese_flag and (not self.obese_sticking_to_treatment_flag) and (not self.depressed_flag):#just obese
            obese_seeking_help_flag=0
            return Obese(self.unique_id, self.model, self.depressed_flag, self.obese_flag, obese_seeking_help_flag)
        if (not self.obese_flag) and self.depressed_flag: # not obese anymore but now depressed:
            depressed_seeking_help_flag=0
            return Depressed(self.unique_id, self.model, self.depressed_flag,self.obese_flag,depressed_seeking_help_flag)
        if (not self.obese_flag) and (not self.depressed_flag):#not obese and not depressed anymore, which probably would never happen
            return Healthy(self.unique_id, self.model, self.depressed_flag, self.obese_flag)
```

```python
class Depressed_Obese(Agent):
    """
    This is the class of obese and depressed population, an obese and depressed person can seek help in depression or obesity or both
    (or not)
    
    attribute:
    unique_id: unique int
    depress_flag: int, value of 1 or 0, indicating if a person is depressed or not
    obese_flag: int, value of 1 or 0, indicating if a person is obese or not
    depressed_seeking_help_flag: int, value of 1 or 0, indicating if this person is seeking help with depression
    obese_seeking_help_flag: int, value of 1 or 0, indicating if this person is seeking help with obesity
    """
    def __init__(self, unique_id, model, depressed_flag,obese_flag,depressed_seeking_help_flag, obese_seeking_help_flag):
        """
        
        """
        super().__init__(unique_id, model)
        #self.breed = 'depressed_and_obese_people'
        self.depressed_flag=depressed_flag
        self.obese_flag=obese_flag
        self.depressed_seeking_help_flag=depressed_seeking_help_flag
        self.obese_seeking_help_flag=obese_seeking_help_flag
        
    def step(self):
        """
        This is the function which would randomly decide if an obese and depressed person is going to seek help in either 
        way (or not) based on the probabilities. If so then it will return the corresponding new type of this person, 
        if not then it will return its old self
        """
        self.depressed_seeking_help_flag=numpy.random.choice(numpy.arange(0,2), p=[(1-depressed_seeking_help_prob), depressed_seeking_help_prob])
        self.obese_seeking_help_flag=numpy.random.choice(numpy.arange(0,2), p=[(1-obese_seeking_help_prob), obese_seeking_help_prob])
        if self.depressed_seeking_help_flag or self.obese_seeking_help_flag:#depressed (seeking help with depression) and obese (and seeking help with obesity)
            depressed_sticking_to_treatment_flag=0
            obese_sticking_to_treatment_flag=0
            return Depressed_Obese_Seeking_Help(self.unique_id, self.model, self.depressed_flag, self.obese_flag,self.depressed_seeking_help_flag, self.obese_seeking_help_flag, depressed_sticking_to_treatment_flag, obese_sticking_to_treatment_flag)
        else:#depressed and obese and not seeking any help
            return self
```

```python
class Depressed_Obese_Seeking_Help(Agent):
    """
   This is the class of obese and depressed and seeking help (in either way or both )population, an obese and depressed 
   and seeking help in depression or obesity or both (or not) person can get rid of obesity (not depression though)
    
    attribute:
    unique_id: unique int
    depress_flag: int, value of 1 or 0, indicating if a person is depressed or not
    obese_flag: int, value of 1 or 0, indicating if a person is obese or not
    depressed_seeking_help_flag: int, value of 1 or 0, indicating if this person is seeking help with depression
    obese_seeking_help_flag: int, value of 1 or 0, indicating if this person is seeking help with obesity
    depressed_sticking_to_treatment_flag: int, value of 1 or 0, indicating if this person is sticking with treatment in depression
    obese_sticking_to_treatment_flag: int, value of 1 or 0, indicating if this person is sticking with treatment in obesity
    """
    def __init__(self, unique_id, model, depressed_flag, obese_flag,depressed_seeking_help_flag, obese_seeking_help_flag, depressed_sticking_to_treatment_flag, obese_sticking_to_treatment_flag):
        """
       
        """
        super().__init__(unique_id, model)
        #self.breed = 'depressed_and_obese_people_seeking_help'
        self.depressed_flag=depressed_flag
        self.obese_flag=obese_flag
        self.depressed_seeking_help_flag=depressed_seeking_help_flag
        self.obese_seeking_help_flag=obese_seeking_help_flag
        self.depressed_sticking_to_treatment_flag=depressed_sticking_to_treatment_flag
        self.obese_sticking_to_treatment_flag=obese_sticking_to_treatment_flag
        
    def step(self):
        """
        This is the function which would randomly decide if an obese and depressed and seeking help person is going to
        seek more help in either way (or not) and get rid of obesity (or not) based on the probabilities. If so then 
        it will return the corresponding new type of this person, if not then it will return its old self
        """
        if self.depressed_seeking_help_flag:
            self.depressed_sticking_to_treatment_flag=numpy.random.choice(numpy.arange(0,2), p=[(1-depressed_sticking_to_treatment_prob), depressed_sticking_to_treatment_prob])
        if self.obese_seeking_help_flag:
            self.obese_sticking_to_treatment_flag=numpy.random.choice(numpy.arange(0,2), p=[(1-obese_sticking_to_treatment_prob), obese_sticking_to_treatment_prob])
        
        if(self.obese_sticking_to_treatment_flag):
            self.obese_flag=numpy.random.choice(numpy.arange(0,2), p=[obese_healthy_prob, (1-obese_healthy_prob)])
        else:
            self.obese_flag=1

        if self.obese_flag and self.obese_sticking_to_treatment_flag and self.depressed_sticking_to_treatment_flag:#obese (but sticking to treatment) and depressed (and sticking to treatment)
            depressed_seeking_help_flag=1
            depressed_sticking_to_treatment_flag=1
            return Depressed_Obese_Seeking_Help(self.unique_id, self.model, self.depressed_flag,self.obese_flag,depressed_seeking_help_flag, self.obese_seeking_help_flag, depressed_sticking_to_treatment_flag, self.obese_sticking_to_treatment_flag)
        if self.obese_flag and self.obese_sticking_to_treatment_flag and (not self.depressed_sticking_to_treatment_flag):#obese and sticking to treatment and depressed but not sticking with treatment
            depressed_seeking_help_flag=0
            depressed_sticking_to_treatment_flag=0
            return Depressed_Obese_Seeking_Help(self.unique_id, self.model, self.depressed_flag,self.obese_flag,depressed_seeking_help_flag, self.obese_seeking_help_flag, depressed_sticking_to_treatment_flag, self.obese_sticking_to_treatment_flag)
        if self.obese_flag and (not self.obese_sticking_to_treatment_flag) and self.depressed_sticking_to_treatment_flag:#obese but not sticking to treatment and depressed (and sticking to treatment)
            obese_seeking_help_flag=0
            obese_sticking_to_treatment_flag=0
            return Depressed_Obese_Seeking_Help(self.unique_id, self.model, self.depressed_flag,self.obese_flag,self.depressed_seeking_help_flag, obese_seeking_help_flag, self.depressed_sticking_to_treatment_flag, obese_sticking_to_treatment_flag)
        if self.obese_flag and (not self.obese_sticking_to_treatment_flag) and (not self.depressed_sticking_to_treatment_flag):#obese and depressed only (not seeking any treatment)
            depressed_seeking_help_flag=0
            obese_seeking_help_flag=0
            return Depressed_Obese(self.unique_id, self.model, self.depressed_flag,self.obese_flag,depressed_seeking_help_flag, obese_seeking_help_flag)
        if (not self.obese_flag) and self.depressed_sticking_to_treatment_flag:#not obese, but depressed and sticking witb treatment:
            return Depressed_Seeking_Help(self.unique_id, self.model, self.depressed_flag, self.obese_flag,self.depressed_sticking_to_treatment_flag)
        if (not self.obese_flag) and (not self.depressed_sticking_to_treatment_flag):#not obese but depressed and not sticking with treatment
            depressed_seeking_help_flag=0
            return Depressed(self.unique_id, self.model, self.depressed_flag,self.obese_flag,depressed_seeking_help_flag)
```
